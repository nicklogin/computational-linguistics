Попробуем выкачать все приключения для начинающих игроков:

```python
url = 'http://www.dndadventure.com/html/adventures/adv1.html'
```

```python
import urllib.request
```

```python
with urllib.request.urlopen(url) as start_page:
    pagetext = start_page.read().decode('utf-8')
```

```python
import re
```

```python
links = re.findall('"http://.*?\.zip"', pagetext)
```

```python
links
```

```python
import zipfile
```

```python
help(zipfile)
```

```python
from io import BytesIO
```

```python
links = list(map(lambda x: x.strip('"'), links))
```

```python
z = zipfile.ZipFile(BytesIO(urllib.request.urlopen(links[0]).read()))
```

```python
z.namelist()
```

```python
z.close()
```

Извлечём все документы:

```python
for link in links:
    z = zipfile.ZipFile(BytesIO(urllib.request.urlopen(link).read()))
    z.extractall(path_to_data)
    z.close()
```

Посмотрим папку, куда извлеклись документы:

```python
import os
```

```python
[i for i in os.listdir(path_to_data) if i.endswith('.doc')]
```

У нас есть файлы в формате txt, pdf, doc и rtf. Pdf будем открывать пакетом tika, txt - стандартными средствами питона, rtf - скриптом extract_rtf <a href="https://gist.githubusercontent.com/gilsondev/7c1d2d753ddb522e7bc22511cfb08676/raw/163ef3ee9893109f10cccf4b51d1d077ada296fa/extract_rtf.py">отсюда</a>

DOC извлечь нормально не получится, поэтому тупо конвертим их в RTF при помощи Ворда (тем более, таких файлов всего 4)

```python
import extract_rtf
```

```python
with open(path_to_data+'Hommlet NPCs.rtf', 'r', encoding='utf-8') as inp:
    rtftext = inp.read()
```

```python
print(extract_rtf.striprtf(rtftext))
```

```python
texts = []
```

Извлечём текст из всех документов:

```python
for i in os.listdir(path_to_data):
    if i.endswith('.txt'):
        with open(path_to_data + i, 'r', encoding='utf-8') as inp:
            texts.append(inp.read())
    elif i.endswith('.rtf'):
        with open(path_to_data + i, 'r', encoding='utf-8') as inp:
            texts.append(extract_rtf.striprtf(inp.read()))
    elif i.endswith('.pdf'):
        texts.append(parser.from_file(path_to_data + i)['content'])
```

```python
len(texts)
```

Всего получилось 27 текстов, посмотрим сколько это токенов:

```python
s = 0
for text in texts:
    s += len(word_tokenize(text))
s
```

403366 токенов - не так уж и мало, почтиполовина реалека.

Сохраним всё это в json:

```python
import json
```

```python
with open('project/processed_data.json', 'w', encoding='utf-8') as outp:
    json.dump(texts, outp, ensure_ascii=False)
```
